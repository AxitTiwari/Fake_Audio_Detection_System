```python
spec = np.abs(librosa.stft(y, hop_length=512))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
plt.title('Spectrogram');
```

```python
mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time');
plt.title('Mel Spectrogram');
plt.colorbar(format='%+2.0f dB');
```

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
real_audio_path = "real_sample.wav"
fake_audio_path = "fake_sample.wav"

real_audio, sr = librosa.load(real_audio_path, sr=16000)
fake_audio, sr = librosa.load(fake_audio_path, sr=16000)
```

```python
def get_mel_spectrogram(audio, sr):
    mel_spec = librosa.feature.melspectrogram(
        y=audio,
        sr=sr,
        n_fft=2048,
        hop_length=512,
        n_mels=128
    )
    
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    return mel_spec_db
```

```python
real_spec = get_mel_spectrogram(real_audio, sr)
fake_spec = get_mel_spectrogram(fake_audio, sr)
```

```python
plt.figure(figsize=(12, 5))

# Real Audio
plt.subplot(1, 2, 1)
librosa.display.specshow(real_spec, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
plt.title("Real Audio Spectrogram")
plt.colorbar(format='%+2.0f dB')

# Fake Audio
plt.subplot(1, 2, 2)
librosa.display.specshow(fake_spec, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
plt.title("Fake Audio Spectrogram")
plt.colorbar(format='%+2.0f dB')

plt.tight_layout()
plt.show()
```

## Real to one folder

```python
import os
import shutil

input_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\real"   
output_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\real"

os.makedirs(output_folder, exist_ok=True)

count = 0

for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith((".wav", ".mp3", ".flac")):
            old_path = os.path.join(root, file)
            
            new_name = f"audio_{count}.wav"   # rename properly
            new_path = os.path.join(output_folder, new_name)
            
            shutil.copy(old_path, new_path)
            count += 1

print("Total files collected:", count)
```

## Fake to one folder

```python
import os
import librosa
import soundfile as sf

# 🔴 CHANGE THIS PATH
input_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\fake\tts-samples-main\mp3\English"

# 🔴 OUTPUT FOLDER
output_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\fake"

os.makedirs(output_folder, exist_ok=True)

count = 0

for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.lower().endswith((".wav", ".mp3", ".flac")):
            file_path = os.path.join(root, file)
            
            try:
                # Load audio and resample to 16kHz
                y, sr = librosa.load(file_path, sr=16000)
                
                # Save as WAV
                new_file = os.path.join(output_folder, f"fake_{count}.wav")
                sf.write(new_file, y, sr)
                
                print(f"Saved: {new_file}")
                count += 1
                
            except Exception as e:
                print(f"Error with {file_path}: {e}")

print("\n✅ Done!")
print("Total fake English audio files:", count)
```

## Data augumentation

```python
import os
import librosa
import soundfile as sf
import numpy as np

input_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\fake"
output_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\fake"

os.makedirs(output_folder, exist_ok=True)

count = 0

for file in os.listdir(input_folder):
    path = os.path.join(input_folder, file)
    
    y, sr = librosa.load(path, sr=16000)
    
    # 1. Original
    sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y, sr)
    count += 1

    # 2. Add noise
    noise = np.random.normal(0, 0.005, y.shape)
    sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y + noise, sr)
    count += 1

    # 3. Pitch shift
    y_pitch = librosa.effects.pitch_shift(y, sr=sr, n_steps=2)
    sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y_pitch, sr)
    count += 1

    # 4. Speed change
    y_fast = librosa.effects.time_stretch(y, rate=1.2)
    sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y_fast, sr)
    count += 1

print("Total augmented fake samples:", count)
```

```python
import os
import librosa
import soundfile as sf
import numpy as np

input_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\fake"
output_folder = r"D:\YEAR-2\SEM IV\audio_dection\raw_data\fake"

os.makedirs(output_folder, exist_ok=True)

count = 0
skipped = 0

for file in os.listdir(input_folder):
    path = os.path.join(input_folder, file)

    try:
        y, sr = librosa.load(path, sr=16000)

        # 1. Original
        sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y, sr)
        count += 1

        # 2. Noise
        noise = np.random.normal(0, 0.005, y.shape)
        sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y + noise, sr)
        count += 1

        # 3. Pitch
        y_pitch = librosa.effects.pitch_shift(y, sr=sr, n_steps=2)
        sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y_pitch, sr)
        count += 1

        # 4. Speed
        y_fast = librosa.effects.time_stretch(y, rate=1.2)
        sf.write(os.path.join(output_folder, f"fake_{count}.wav"), y_fast, sr)
        count += 1

    except Exception as e:
        print(f"❌ Skipped: {file}")
        skipped += 1

print("\n✅ Done!")
print("Generated files:", count)
print("Skipped files:", skipped)
```
